```python
"""Import libraries for make some calculations with data and understand data."""
import pandas as pd
import numpy as np
```

```python
"""With pandas library read csv file for importing python."""
df = pd.read_csv('titanic.csv')
```

```python
"""Show first 5 row in dataframe."""
df.head()
```

```python
"""This is how you drop column
    axis = 1 means column
    axis = 0 means row."""
df.drop(labels = ["SibSp"],axis=1, inplace=True)

```

```python
df
```

```python
""" This is howyou drop only female rows in dataframe :"""
df[df["Sex"] == "female"]
df.drop(labels = df[df["Sex"] == "female"].index)
```

```python
"""In dataframe may there are duplicates which you dont want in your dataframe. 
This is how you drop duplicates in your dataframe"""
df.drop_duplicates()
```

```python
"""Compare and taking some information with group column"""
df.groupby(by=["Sex"]).mean()
```

```python
df.groupby(by=["Sex"]).count()
```

```python
"""Compare and taking some information with group two column"""
df.groupby(by=["Sex","Survived"]).count()
```

```python
df.groupby(by=["Sex","Survived"]).mean()
```
